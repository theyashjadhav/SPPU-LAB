```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 5*x + 6
```

```python
def df(x):
    return 2*x + 5
```

```python
def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x
    x_history = [x]

    for i in range(num_iterations):
        gradient = df(x)
        x = x - learning_rate * gradient
        x_history.append(x)

    return x, x_history
```

```python
initial_x = -10
learning_rate = 0.1
num_iterations = 10
```

```python
x, x_history = gradient_descent(initial_x, learning_rate, num_iterations)
print("Local minimum: {:.2f}".format(x))
```

```python
for i, x_val in enumerate(x_history):
    print(f"Iteration {i+1}: x = {x_val:.2f}, f(x) = {f(x_val):.2f}")
```

```python
x_vals = np.linspace(-11, 5, 90)
y_vals = f(x_vals)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^2 + 5x + 6')
plt.scatter(x_history, [f(x) for x in x_history], color='red')

for i in range(1, len(x_history)):
    plt.plot([x_history[i-1], x_history[i]], [f(x_history[i-1]), f(x_history[i])], color='blue', linestyle='--')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
